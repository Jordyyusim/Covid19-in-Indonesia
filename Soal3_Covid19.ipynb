```python
import requests
import json
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium as fo
```

```python
datageo = pd.read_json('http://raw.githubusercontent.com/LintangWisesa/Indonesia-Covid19-Maps/master/data/gps_indonesia.json')
datageolat = datageo[['latitude']]

datageolong = datageo[['longitude']]
datageologo = datageo[['logo']]
datageoprovinsi = datageo[['Provinsi']]
datageo['latitude'][0]
datageo['logo'][10]
```

```python
d = pd.read_html('https://en.wikipedia.org/wiki/2020_coronavirus_pandemic_in_Indonesia')
d = d[3].loc[0:34]
d.columns = [' '] * len(d.columns)
d.dropna(axis=1, inplace=True)
d.columns = ['Provinsi', 'Confirmed', 'Recovered', 'Deaths', 'Active']
# d = d[['Confirmed', 'Recovered', 'Deaths']]
d_confirmed = 'Confirmed:'+ d['Confirmed']
d_recovered = 'Recovered:' + d['Recovered']
d_deaths = 'Deaths:' + d['Deaths']
d_confirmed[0]
d_recovered[0]

```

```python
Indonesia = fo.Map(
    Location = [-1.6538255, 119.1944534],
    zoom_start = 5.21)

for i in range(34):
    fo.Marker([datageo['latitude'][i], datageo['longitude'][i]],
              tooltip = datageo['Provinsi'][i],
              popup = [d_confirmed[i], d_recovered[i], d_deaths[i]],
              icon = fo.features.CustomIcon(datageo['logo'][i])).add_to(Indonesia)
              
Indonesia
# Indonesia.save('Indonesia.html')
              
    
```
