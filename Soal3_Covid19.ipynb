```python
import requests
import json
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium as fo
```

```python
datageo = pd.read_json('http://raw.githubusercontent.com/LintangWisesa/Indonesia-Covid19-Maps/master/data/gps_indonesia.json')
datageolat = datageo[['latitude']]

datageolong = datageo[['longitude']]
datageologo = datageo[['logo']]
datageoprovinsi = datageo[['Provinsi']]
datageoprovinsi
```

```python
d = pd.read_html('https://en.wikipedia.org/wiki/2020_coronavirus_pandemic_in_Indonesia')
d = d[3]
d.columns = [' '] * len(d.columns)
d.dropna(axis=1, inplace=True)
d.columns = ['Provinsi', 'Confirmed', 'Recovered', 'Deaths', 'Active']
d = d[['Confirmed', 'Recovered', 'Deaths']]
d

```

```python
Indonesia = fo.Map(
    Location = [-1.6538255, 119.1944534],
    zoom_start = 5.21)

for i in range(34):
    fo.Marker([datageolat.iloc[i], datageolong.iloc[i]],
              tooltip = datageoprovinsi.iloc[i],
              popup = d.iloc[i]).add_to(Indonesia)
#               icon = fo.features.CustomIcon(datageologo.iloc[i])).add_to(Indonesia)
              
Indonesia
              
              
    
```
